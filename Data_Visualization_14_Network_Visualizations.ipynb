```python
import networkx as nx
import matplotlib.pyplot as plt
```

BUAT NODE MASING-MASING KOTA <br>
Sumber Gambar = https://radarsukabumi.com/wp-content/uploads/2019/10/peta-jabar.jpg

```python
G=nx.Graph()
G.add_node('Depok',pos=(1,10))
G.add_node('Cibinong',pos=(1,9))
G.add_node('Bogor',pos=(1,8))
G.add_node('Bekasi',pos=(2,11))
G.add_node('Sukabumi',pos=(2,4))
G.add_node('Cianjur',pos=(3,6))
G.add_node('Karawang',pos=(4,10))
G.add_node('Purwakarta',pos=(5,8))
G.add_node('Cimahi',pos=(5,5))
G.add_node('Soreang',pos=(5,3))
G.add_node('Bandung',pos=(6,5))
G.add_node('Subang',pos=(7,8))
G.add_node('Sumedang',pos=(8,6))
G.add_node('Garut',pos=(8,2))
G.add_node('Majalengka',pos=(9,6))
G.add_node('Tasikmalaya',pos=(9,1))
G.add_node('Indramayu',pos=(10,10))
G.add_node('Kuningan',pos=(11,4))
G.add_node('Ciamis',pos=(11,1))
G.add_node('Cirebon',pos=(12,7))
G.add_edges_from([('Depok', 'Cibinong'), ('Cibinong', 'Bogor'), ('Bekasi', 'Karawang'), ('Karawang', 'Purwakarta'), ('Karawang', 'Indramayu'), ('Bogor', 'Cianjur'), ('Cianjur', 'Sukabumi'), ('Bogor', 'Sukabumi'), ('Cianjur', 'Cimahi'), ('Cimahi', 'Bandung'), ('Purwakarta', 'Bandung'), ('Purwakarta', 'Subang'), ('Subang', 'Bandung'), ('Soreang', 'Bandung'), ('Sumedang', 'Bandung'), ('Garut', 'Bandung'), ('Tasikmalaya', 'Bandung'), ('Ciamis', 'Tasikmalaya'), ('Kuningan', 'Majalengka'), ('Ciamis', 'Sumedang'), ('Ciamis', 'Kuningan'), ('Ciamis', 'Majalengka'), ('Cirebon', 'Kuningan'), ('Cirebon', 'Majalengka'), ('Cirebon', 'Sumedang'), ('Cirebon', 'Indramayu'), ('Indramayu', 'Majalengka'), ('Indramayu', 'Sumedang'), ('Subang', 'Indramayu')])
pos=nx.get_node_attributes(G,'pos')
plt.figure(figsize=(15,10))
nx.draw(G, pos, with_labels = True)
```
