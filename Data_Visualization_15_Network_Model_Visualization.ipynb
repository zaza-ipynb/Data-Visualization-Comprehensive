# <center>Maximal cliques</center>

## Maximal cliques

- Definition: a clique that, when extended by one node is no longer a clique



- Applications: community finding



## Communities

- Find cliques
- Find unions of cliques



## NetworkX API

- find_cliques finds all maximal cliques

## Maximal cliques

```python
import networkx as nx
G = nx.barbell_graph(m1=5, m2=1)
nx.draw(G, with_labels = True)
plt.show()
```

```python
nx.find_cliques(G)
```

```python
list(nx.find_cliques(G))
```

# <center>Subgraphs</center>

## Subgraphs

- Visualize portions of a large graph
    - Paths
    - Communities/cliques
    - Degrees of separation from a node

```python
import networkx as nx
G = nx.erdos_renyi_graph(n=20, p=0.2)
nx.draw(G, with_labels = True)
plt.show()
```

```python
G.nodes()
```

```python
nodes = list(G.neighbors(4))
```

```python
list(nodes)
```

```python
nodes.append(4)
nodes
```

```python
G_four = G.subgraph(nodes)
```

```python
G_four.edges()
```

```python
nx.draw(G_four, with_labels=True)
plt.show()
```
